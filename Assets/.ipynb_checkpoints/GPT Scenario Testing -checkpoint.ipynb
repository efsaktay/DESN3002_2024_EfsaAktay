```python
import pandas as pd
import numpy as np
import os
import torch
th = torch
import numpy as np
import pandas as pd
import yfinance as yf
import pandas as pd
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler

# Setting the random seed for reproducibility
np.random.seed(42)

# Number of trading days
n_days = 1047

# Helper function to generate a tweet
# Helper function to generate a tweet
def generate_tweet(sentiment):
    products = ["latest model", "innovative service", "cutting-edge solution", "revolutionary app", "next-gen technology"]
    achievements = ["has just been launched", "is breaking new ground", "has shattered expectations", "is a game-changer"]
    actions = ["pushing boundaries", "leading the market", "driving success", "setting new standards", "capturing new territories"]
    challenges = ["economic downturns", "competitive pressures", "tightening regulations", "supply chain disruptions"]
    strategies = ["pivot our focus", "leverage new opportunities", "double down on our core strengths", "diversify our approaches"]
    updates = ["earnings call", "stakeholder meeting", "press release", "product reveal", "partnership announcement"]
    adjectives = ["challenging", "rewarding", "transformative", "significant", "critical"]
    outcomes = ["overcome these challenges", "build on these successes", "continue our growth", "maintain our market position"]
    events = ["in our latest webinar", "during the recent industry conference", "in our annual report"]

    if sentiment == 'positive':
        choice = np.random.choice([
            f"Our {np.random.choice(products)} {np.random.choice(achievements)}! A true testament to our commitment to excellence. #LeadingEdge",
            f"Exciting times! The launch of our {np.random.choice(products)} has officially set a new standard. #InnovationLeadership",
            f"We're proud to announce that our {np.random.choice(products)} is now {np.random.choice(achievements)}. Thanks to everyone involved! #TeamEffort",
            f"Breaking news: Our {np.random.choice(products)} has revolutionized the industry by {np.random.choice(actions)}. #MarketLeader",
            f"Today marks a historic moment as our {np.random.choice(products)} achieves {np.random.choice(achievements)}. #PioneeringSuccess"
        ], 1)[0]

    elif sentiment == 'negative':
        choice = np.random.choice([
            f"Challenging times ahead as we face {np.random.choice(challenges)}. We're strategizing to {np.random.choice(strategies)}. #StayStrong",
            f"Due to {np.random.choice(challenges)}, we have to {np.random.choice(strategies)}. Your support is vital. #Unity",
            f"Adapting to {np.random.choice(challenges)} is tough, but necessary. We're committed to {np.random.choice(outcomes)}. #Resilience",
            f"The recent {np.random.choice(challenges)} require us to {np.random.choice(strategies)}. We believe in our path forward. #Determination",
            f"While facing {np.random.choice(challenges)}, we're taking steps to {np.random.choice(strategies)}. It's a critical time. #CriticalTimes"
        ], 1)[0]

    elif sentiment == 'neutral':
        choice = np.random.choice([
            f"Don't miss our upcoming {np.random.choice(updates)} {np.random.choice(events)}. It's going to be informative. #StayTuned",
            f"Just a reminder about our {np.random.choice(updates)} tomorrow. We look forward to your participation. #EngageWithUs",
            f"Tune in for our {np.random.choice(updates)} this week. Exciting updates to share! #CorporateUpdate",
            f"We're excited to host the {np.random.choice(updates)} next month. Join us to learn more about our progress. #OpenForAll",
            f"Mark your calendars! Our {np.random.choice(updates)} is scheduled for next Friday. We have important news to share. #Announcement"
        ], 1)[0]

    return choice

# Generate tweets and sentiments for each CEO over the days
dates = pd.date_range(start='2020-01-01', periods=n_days, freq='B')  # Business days

# Dataframe to store CEO tweets and sentiments
ceo_tweet_data = {
    'Date': dates,
    'CEO1_tweet': [None]*n_days,
    'CEO1_sentiment': [None]*n_days,
    'CEO2_tweet': [None]*n_days,
    'CEO2_sentiment': [None]*n_days,
    'CEO3_tweet': [None]*n_days,
    'CEO3_sentiment': [None]*n_days
}

# Simulate tweeting behavior for 3 CEOs with different frequencies and sentiments
for i, date in enumerate(dates):
    if np.random.rand() < 0.7:  # CEO 1 tweets most frequently
        sentiment = np.random.choice(['positive', 'negative', 'neutral'], p=[0.15, 0.7, 0.15])
        ceo_tweet_data['CEO1_tweet'][i] = generate_tweet(sentiment)
        ceo_tweet_data['CEO1_sentiment'][i] = sentiment

    if np.random.rand() < 0.5:  # CEO 2 tweets less frequently
        sentiment = np.random.choice(['positive', 'negative', 'neutral'], p=[0.7, 0.15, 0.15])
        ceo_tweet_data['CEO2_tweet'][i] = generate_tweet(sentiment)
        ceo_tweet_data['CEO2_sentiment'][i] = sentiment
        
    if np.random.rand() < 0.3:  # CEO 3 tweets least frequently
        sentiment = np.random.choice(['neutral', 'positive', 'negative'], p=[0.7, 0.15, 0.15])
        ceo_tweet_data['CEO3_tweet'][i] = generate_tweet(sentiment)
        ceo_tweet_data['CEO3_sentiment'][i] = sentiment

# Create a DataFrame from the tweet data
df_ceo_tweets = pd.DataFrame(ceo_tweet_data)

# Display the first few rows of the dataframe
print(df_ceo_tweets.head())

# Save the dataframe to a CSV file
df_ceo_tweets.to_csv('ceo_tweets_with_sentiment.csv', index=False)

# Print summary
print(f"Generated DataFrame with {len(df_ceo_tweets)} rows.")
```

```python
import matplotlib.pyplot as plt
import pickle

# Load the serialized data
loaded_dfs = pickle.load(open('scaled_stocks_data.pkl', 'rb'))

# Normalize the data
normalized_dfs = {}
for stock, df in loaded_dfs.items():
    normalized_df = df.div(df.iloc[0])
    normalized_dfs[stock] = normalized_df

# Set up the plotting style
plt.style.use('seaborn-darkgrid')

# Create a figure with multiple subplots
fig, axes = plt.subplots(nrows=len(normalized_dfs), figsize=(12, 8), dpi=100, sharex=True)

# Ensure 'axes' is always an array even if there's only one subplot
if len(normalized_dfs) == 1:
    axes = [axes]

# Plot each normalized DataFrame
for ax, (stock, df) in zip(axes, normalized_dfs.items()):
    # Plot the close price of the stock
    ax.plot(df['close'], label=f'{stock} Closing Price (Normalized)', linewidth=2)
    
    # Plot the index prices
    for index in ['Index 1', 'Index 2', 'Index 3']:
        ax.plot(df[index], label=f'{index} Price (Normalized)', linestyle='--')
    
    ax.set_title(f'{stock} Prices and Indexes (Normalized)')
    ax.set_ylabel('Normalized Price')
    ax.legend()

# Set common labels
axes[-1].set_xlabel('Date')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
